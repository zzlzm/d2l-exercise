```python
import requests
from lxml import etree
from docx import Document
```

```python
def get_art(url):
    r = requests.get(url, allow_redirects=True)
    dom = etree.HTML(r.content)
    header = dom.xpath('//h2/text()')[0].replace('/', '')
    print(header)
    print(url)
#     article_body = dom.xpath('//div[@id="articlebody"]//p')
    article_body = dom.xpath('//div[@id="sina_keyword_ad_area2"]')
    text_list = []
    for p in article_body:
        for t in p.itertext():
            text_list.append(t)
    text = ''.join(text_list)
#     print(text)
    doc = Document()
    doc.add_paragraph(url)
    doc.add_paragraph(text)
    doc.save('./data/' + header + '.docx')
    
```

```python
def fetch_articles():
    urls = ['http://blog.sina.com.cn/s/articlelist_1243629662_1_1.html',
           'http://blog.sina.com.cn/s/articlelist_1243629662_12_1.html',
           'http://blog.sina.com.cn/s/articlelist_1243629662_16_1.html']
    for url in urls:
        r = requests.get(url, allow_redirects=True)
        dom = etree.HTML(r.content)
        a_tags = dom.xpath('//span[@class="atc_title"]/a')
        art_list = [ele.attrib['href'] for ele in a_tags]
        for art in art_list:
            get_art(art)
def test():
    url = 'http://blog.sina.com.cn/s/articlelist_1243629662_1_1.html'
    r = requests.get(url, allow_redirects=True)
    dom = etree.HTML(r.content)
    a_tags = dom.xpath('//span[@class="atc_title"]/a')
    art_list = [ele.attrib['href'] for ele in a_tags]
    get_art(art_list[0])
# test()
fetch_articles()
```
